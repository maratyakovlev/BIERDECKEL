During the analysis of the Dataset I noticed that in some cases renewal date is equal or less than order date.

```python
orders_df['check'] = orders_df['OrderDate'] < orders_df['Contact: TotalTheo Renewal Date']
display(orders_df.query('check == False'))
```

It looks like a database error. For the sake of analysis I would assume that renewal date should be the last order date plus 90 days (assuming that most of the orders are for quarter subscription).

```python
#Function below fixes renewal dates. If renewal date is equal or less than order date,
#it replaces renewal date with the latest order date + 90 days.

def renewal_date_check(order_date,renewal_date,client_id,df):
    last_order_date = df[df['Contact Id'] == client_id ]['OrderDate'].max()
    if renewal_date <= order_date:
        renewal_date = last_order_date + pd.Timedelta(days=90)
        return renewal_date
    else:
        return renewal_date
```

```python
#Let's apply this function to the Dataset.
orders_df['Contact: TotalTheo Renewal Date'] = orders_df.apply(lambda row: renewal_date_check(row[3],row[6],row[1],orders_df), axis=1)
```

In the dataset we have info about orders made from 09/02/2021. At the same time we have clients with first purchase from 2016. Let's try to "restore" info about previous purchases of these clients to perform further analysis of their lifetime behavior.
If the first purchase of the client is before 09/02/2021 we would add additional rows with previous purchases assuming quarter renewal.

Since code below requires a lot of time (at least on my machine) - let's drop columns unnecessary for further analysis: Id, OrderId, OrderTotal, Invoice Id.

```python
orders_df.drop(labels=['Id','OrderId','OrderTotal','Invoice id'], axis=1, inplace=True)
```

```python
display(orders_df.head())
```

```python
df_ind = range(0,len(orders_df.index))

for row in df_ind:
    purchase_date = orders_df.iloc[row,3]
    while (purchase_date+pd.Timedelta(days=90)) < orders_df.iloc[row,1]:
        first_purchase_row = pd.DataFrame([[orders_df.iloc[row,0],purchase_date,None,orders_df.iloc[row,7]]], columns = orders_df.columns)
        orders_df = pd.concat([orders_df,first_purchase_row],ignore_index=True)
        purchase_row = pd.DataFrame([[None,orders_df.iloc[row,1],None,purchase_date+pd.Timedelta(days=90),None,None,None,orders_df.iloc[row,7]]], columns = orders_df.columns)
        orders_df = pd.concat([orders_df,purchase_row],ignore_index=True)
        purchase_date = purchase_date + pd.Timedelta(days=90)

```

Let's create quarter cohorts of clients based on First purchase date and then look at quarter renewal orders per quarter.

For columns Order date and First Purchase Date let's create additional columns with quarter-year info.

```python
orders_df['fp_quarter'] = orders_df['Contact: First Purchase Date'].dt.quarter
orders_df['fp_quarter'] = orders_df['fp_quarter'].apply(str)
orders_df['fp_year'] = orders_df['Contact: First Purchase Date'].dt.year
orders_df['fp_year'] = orders_df['fp_year'].apply(str)
orders_df['fp_Q-Y'] = orders_df['fp_year'].astype(str)+'-'+orders_df['fp_quarter'].astype(str)

orders_df['order_quarter'] = orders_df['OrderDate'].dt.quarter
orders_df['order_quarter'] = orders_df['order_quarter'].apply(str)
orders_df['order_year'] = orders_df['OrderDate'].dt.year
orders_df['order_year'] = orders_df['order_year'].apply(str)
orders_df['order_Q-Y'] = orders_df['order_year'].astype(str)+'-'+orders_df['order_quarter'].astype(str)
```
